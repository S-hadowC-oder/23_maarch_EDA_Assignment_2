# Amazon Sales Data

## Description:
### This dataset contains information on 1K+ Amazon products, including their ratings, reviews, and other details.

## Features:
### product_id: Unique identifier for each product

### product_name: Name of the product

### category: Category of the product

### discounted_price: Discounted price of the product

### actual_price: Actual price of the product

### discount_percentage: Percentage of discount for the product

### rating: Rating of the product (1-5)

### rating_count: Number of people who voted for the Amazon rating

### about_product: Description about the product

### user_id: ID of the user who wrote the review

### user_name: Name of the user who wrote the review

### review_id: ID of the user review

### review_title: Short review

### review_content: Long review

### img_link: Image link of the product

### product_link: Official website link of the product

```python
df['rating_count'] = pd.to_numeric(df['rating_count'],errors='coerce')

# Drop rows with missing values in 'category' or 'rating_count' columns
df = df.dropna(subset=['category', 'rating_count'])

# Convert 'rating_count' to integer (if needed)
# amazon_data['rating_count'] = amazon_data['rating_count'].astype(int)

# Find the top product by rating_count in each category
top_products_by_category = df.loc[df.groupby('category')['rating_count'].idxmax()]

# Display the result
print(top_products_by_category[['category', 'product_name', 'rating_count']])
```

```python
#Droping the rows with missing values in discounted_price or actual_price
df = df.dropna(subset=['discounted_price','actual_price'])
```

```python
plt.figure(figsize=(12, 6))
plt.hist(df['discounted_price'], bins=30, alpha=0.5, label='Discounted Price') #bins = intervals, alpha = transparency
plt.hist(df['actual_price'], bins=30, alpha=0.5, label='Actual Price')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Distribution of Discounted Prices vs. Actual Prices')
plt.legend()
plt.show()
```

```python
#4 How does the average discounted percentage vary accross categories
```

```python
df = df.dropna(subset=['discount_percentage'])

# Convert 'discount_percentage' to numeric if necessary
df['discount_percentage'] = pd.to_numeric(df['discount_percentage'], errors='coerce')

# Calculate the average discount percentage by category
avg_discount_by_category = df.groupby('category')['discount_percentage'].mean().reset_index()

# Display the result
print(avg_discount_by_category)
```

```python
#5 What are the most popular product names
# Drop rows with missing values in 'product_name' and 'rating_count'
df = df.dropna(subset=['product_name', 'rating_count'])

# Convert 'rating_count' to numeric if needed
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')

# Sort the data by 'rating_count' in descending order to find the most popular products
most_popular_products = df.sort_values(by='rating_count', ascending=False).head(10)

# Display the top 10 most popular products
print(most_popular_products[['product_name', 'rating_count']])
```

```python
#6.What are the most popular product keywords
from collections import Counter

# Drop missing values and split all product names into individual words
keywords = ' '.join(df['product_name'].dropna()).split()

# Count the occurrences of each word
keyword_counts = Counter(keywords)

# Display the 10 most common keywords
print(keyword_counts.most_common(10))
```

```python
#7. what are the most popular product reviews
# Sort by 'rating_count' to find popular reviews
popular_reviews = df[['review_title', 'review_content', 'rating_count']].sort_values(by='rating_count', ascending=False).dropna().head(10)

popular_reviews
```

```python
#8.what is the correlation between discounted_price and rating?
# Ensure 'discounted_price' and 'rating' are numeric
df['discounted_price'] = pd.to_numeric(df['discounted_price'], errors='coerce')
# df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Drop rows with missing values
# df = df.dropna(subset=['discounted_price', 'rating'])

# Calculate correlation
correlation = df['discounted_price'].corr(df['rating'])

print("Correlation between discounted price and rating:", correlation)

```

```python
#9. What are the top 5 categories based on ratings
# Group by category and calculate the mean rating
df['rating'] = pd.to_numeric(df['rating'],errors='coerce')
top_categories_by_rating = df.groupby('category')['rating'].mean().nlargest(5).reset_index()

# Display result
print(top_categories_by_rating)

```
